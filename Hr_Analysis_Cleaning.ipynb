## 🔧 Fix Data Types
Convert date columns from `object` to `datetime`, and fix `ManagerID` to integer.

```python
# Fix DOB - handle 2-digit years correctly (e.g. '75' → 1975 not 2075)
df['DOB'] = pd.to_datetime(df['DOB'], format='%m/%d/%y')
df['DOB'] = df['DOB'].apply(lambda x: x.replace(year=x.year - 100) if x.year > 2005 else x)

# Convert remaining date columns
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['LastPerformanceReview_Date'] = pd.to_datetime(df['LastPerformanceReview_Date'])
```

```python
df['ManagerID'] = df['ManagerID'].astype('Int64')
```

## 🧹 Handle Termination Columns
- Replace `N/A-StillEmployed` in `TermReason` with a cleaner label
- Verify logic: `Termd=0` should have no termination date, `Termd=1` should have one
- Fill missing `DateofTermination` for active employees with the last known date in the dataset

```python
df["TermReason"] = df["TermReason"].replace("N/A-StillEmployed", "Still Employed")
```

```python
df[["Employee_Name", "DateofTermination", "Termd", "TermReason"]]
```

```python
# Validate: all active employees (Termd=0) should have no termination date
(df[df["Termd"] == 0]["DateofTermination"].isna()).all()
```

```python
# Validate: no terminated employees (Termd=1) should have a missing termination date
(df[df["Termd"] == 1]["DateofTermination"].isna()).any()
```

```python
# Convert DateofTermination to datetime first
df["DateofTermination"] = pd.to_datetime(df["DateofTermination"])
```

```python
# Check last known date in dataset to use as fill value for active employees
print(df["LastPerformanceReview_Date"].max())
```

```python
# Fill missing DateofTermination for active employees with dataset's last known date
df["DateofTermination"] = df["DateofTermination"].fillna(pd.Timestamp("2019-02-28"))
```

## ⚙️ Feature Engineering
Create new columns to enrich the dataset for analysis:
- `Age` — employee age based on DOB vs dataset reference date (2019-02-28)
- `AgeGroup` — categorized age buckets
- `TenureYears` — years of service (rounded to 2 decimal places)
- `IsActive` — readable employment status
- `DaysSinceReview` — days since last performance review

```python
# Age as of dataset reference date (2019-02-28)
df['Age'] = (pd.Timestamp('2019-02-28') - df['DOB']).dt.days // 365
```

```python
# Age group buckets
df['AgeGroup'] = df['Age'].apply(lambda x:
    'Under 30' if x < 30 else
    '30s' if x < 40 else
    '40s' if x < 50 else
    '50s' if x < 60 else
    '60+'
)
```

```python
# Years of service
df['TenureYears'] = round((df['DateofTermination'] - df['DateofHire']).dt.days / 365, 2)
```

```python
# Readable employment status
df['IsActive'] = df['Termd'].apply(lambda x: 'Active' if x == 0 else 'Terminated')
```

```python
# Days since last performance review
df['DaysSinceReview'] = (pd.Timestamp('2019-02-28') - df['LastPerformanceReview_Date']).dt.days
```
